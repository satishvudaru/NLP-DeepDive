```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
p_stemmer = PorterStemmer()
```

```python
words =['run','runner','runs','easily','fairly']
```

```python
for word in words:
    print(word + '----->' + p_stemmer.stem(word))
```

```python
from nltk.stem.snowball import SnowballStemmer
```

```python
s_stemmer = SnowballStemmer(language='english')

for word in words:
    print(word+'--->'+s_stemmer.stem(word))
```

```python
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import spacy
nltk.download('wordnet')
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
words = ["running", "studies", "better", "went", "mice", "children"]

print("NLTK Stemming vs NLTK Lemmatization:\n")

for word in words:
    stem = stemmer.stem(word)
    lemma = lemmatizer.lemmatize(word)
    print(f"{word:10} | Stem: {stem:10} | Lemma: {lemma}")
```

```python
import spacy
text = "He studies hard and studied yesterday. Now he is studying."
nlp =spacy.load("en_core_web_sm") 
doc = nlp(text)

print("Lemmatized tokens:")
print([token.lemma_ for token in doc if token.is_alpha])

print("\nStemmed tokens:")
print([stemmer.stem(word) for word in text.split()])

print (f'********************Another Example**************')
text = "I am meeting him at the meeting"

doc = nlp(text)

print("Word | POS | Lemma")
print("---------------------")

for token in doc:
    print(f"{token.text:10} | {token.pos_:6} | {token.lemma_}")
```
