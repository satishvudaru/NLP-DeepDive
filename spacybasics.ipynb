```python
import spacy
```

```python
nlp =spacy.load("en_core_web_sm") 
```

```python
doc=nlp(u'Tesla is looking to buy a U.S. startup for $6 million')
```

```python
for token in doc:
    print(token.text,token.pos,token.dep_)
```

```python
nlp.pipeline
```

```python
doc2=nlp('Tesla isn''t looking at 10 startups anymore.')
for token in doc2:
        print(token.text,token.pos)
```

```python
doc2[8].is_alpha
```
