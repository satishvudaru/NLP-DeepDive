```python

mypdffile=open('Resources/US_Declaration.pdf',mode='rb')
pdf_reader=PyPDF2.PdfReader(mypdffile)
print(len(pdf_reader.pages))
```

```python
page_one=pdf_reader.pages[0]
print(page_one.extract_text())
```

```python
pdf_writer = PyPDF2.PdfWriter()
pdf_writer.add_page(page_one)
pdf_output = open('BRANDNEWPDF.pdf','wb')
pdf_output.close()
```
